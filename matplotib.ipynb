```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
## Scatter
```

```python
x=np.arange(1,6)
y=np.arange(10,15)
print(x)
print(y)
plt.scatter(x,y)
plt.show()
```

```python
x=np.arange(1,6)
y=np.arange(10,15)
print(x)
print()
print(y)
plt.scatter(x,y,color='r',marker='+')
plt.show()
```

```python
x=np.arange(1,25,3)
y=[9,6,7,4,65,3,10,11]
print(x)
print()
print(y)
print('____________visualization_____________')
plt.scatter(x,y,color='r',marker='+')
plt.show()
```

```python
## plot
```

```python
x=np.arange(1,25,3)
y=[9,6,7,4,65,3,10,11]
print(x)
print()
print(y)
plt.title('Morning Batch')
plt.xlabel('values of x')
plt.ylabel('values of y')
print('__________visualization________')
plt.plot(x,y,color='g',marker='+')
plt.show()
```

```python
## Legend
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r')
plt.plot(co,b,color='g')
plt.plot(co,c,color='b')
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='g',label='co and b')
plt.plot(co,c,color= 'b',label='co and c')
plt.legend(loc=0)
plt.show()
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='g',label='co and b')
plt.plot(co,c,color='b',label='co and c')
plt.legend(loc=0,shadow=True,fontsize='large')
plt.show()
```

```python
## Bar plot 
```

```python
courses=['python','java','Netc#','Go']
revenue=[900,1660,850,460]
```

```python
len(courses)
```

```python
a=np.arange(len(courses))
a
```

```python
plt.bar(a,revenue,label='Revenue data ',color='g')
plt.title('BAR_PLOT')
plt.xticks(a,courses)
plt.ylabel('revenue')
plt.legend(loc=0)
plt.show()
```

```python
plt.barh(a,revenue,label='Revenue data',color='g')
plt.title('BAR PLOT')
plt.yticks(a,courses)
plt.xlabel('revenue')
plt.legend(loc=0)
plt.show()
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
plt.subplot(2,1,2)
plt.plot(co,a,color='r',label='co and a')
plt.subplot(2,1,2)
plt.plot(co,b,color='g',label='co and b')
plt.legend(loc=0)
```

```python
## Pie chart
```

```python
x=[40,30,20,10,1]
y=['java','python','c#','PHP','c++']
plt.figure(figsize=(7,7))
plt.pie(x,labels=y)
```

```python
x=[40,30,20,10,1]
y=['java','python','c#','PHP','c++']
plt.figure(figsize=(7,7))
plt.pie(x,labels=y,shadow=True,autopct='%1.1f%%')
```
